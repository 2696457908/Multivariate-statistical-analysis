```python
import rpy2
import rpy2.robjects as robjects
from rpy2.robjects.packages import importr
print(rpy2.__version__)
```

```python
# 访问R实例
pi = robjects.r('pi')
print(type(pi))
print(pi[0])
```

```python
rscript = '''
m = lm(Sepal.Length~Sepal.Width, data = iris)
'''
pymodel = robjects.r(rscript)
res = robjects.r('summary(m)$residuals')
# 输出的是R的格式
res
```

```python
# 将R格式转换为python格式
list(res)
```
